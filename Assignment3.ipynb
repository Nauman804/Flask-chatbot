#  Flask Chatbot

```python

from flask import Flask, render_template, request, jsonify
import datetime

app = Flask(__name__)


def generate_reply(message):
    msg = message.lower()

    if any(word in msg for word in ["hello", "hi", "salam", "hey"]):
        return "Hello! How can I assist you today?"

    if "how are you" in msg:
        return "I'm doing great! Thanks for asking. What about you?"

    if "your name" in msg:
        return "My name is Flask-AI Chatbot. I was created by   Gokboru"

    if "thank" in msg:
        return "You're welcome! Happy to help "

    if "time" in msg:
        now = datetime.datetime.now().strftime("%I:%M %p")
        return f"The current time is {now}."

    if "date" in msg:
        today = datetime.datetime.now().strftime("%d %B %Y")
        return f"Today's date is {today}."

    if "weather" in msg:
        return "I cannot check live weather, but usually it's sunny in Pakistan "

    if "python" in msg:
        return "Python is a powerful programming language. Do you want loops, functions, or file handling help?"

    if "flask" in msg:
        return "Flask is a lightweight web framework for Python. Simple and powerful!"

    if "html" in msg:
        return "HTML is used to create webpage structure. Tags like <h1>, <p>, <div> are important."

    if "assignment" in msg:
        return "Assignments help you improve your skills. I can help you understand them!"

    if "study" in msg or "exam" in msg:
        return "Study smart, not hard! Let me know if you need help with a topic."

    if "joke" in msg:
        return "Why don’t programmers like nature? Too many bugs"

    if "bye" in msg:
        return "Goodbye! Have a great day "

    return ("Sorry, I didn’t understand that.")

@app.route("/")
def home():
    return "Chatbot is running"

@app.route("/chat", methods=["POST"])
def chat():
    user_msg = request.json["message"]
    bot_reply = generate_reply(user_msg)
    return jsonify({"reply": bot_reply})

```
